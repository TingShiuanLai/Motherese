```python
import os
import soundfile as sf

# Your small test folder
chapter_dir = r"Data\Train\librispeech_asr\dev-clean\84\121123"

# Where you want to save the converted wav files
output_chapter_dir = r"Data\Train\librispeech_asr\dev-clean\converted_wav\84\121123"
os.makedirs(output_chapter_dir, exist_ok=True)

# Loop over and convert each .flac
for file in os.listdir(chapter_dir):
    if file.endswith(".flac"):
        flac_path = os.path.join(chapter_dir, file)
        wav_path = os.path.join(output_chapter_dir, file.replace(".flac", ".wav"))

        audio_data, sample_rate = sf.read(flac_path)
        sf.write(wav_path, audio_data, sample_rate)

        print(f"Converted: {flac_path} → {wav_path}")
```

```python
# import os

# # Updated to point to your new .wav directory
# base_wav_dir = "Data\Train\librispeech_asr\dev-clean\converted_wav\84\121123"
# base_transcript_dir = "Data\Train\librispeech_asr\dev-clean\84\121123"

# data = []

# for root, _, files in os.walk(base_wav_dir):
#     for file in files:
#         if file.endswith(".wav"):
#             wav_path = os.path.join(root, file)

#             # Get file ID like 84-121123-0000
#             rel_folder = os.path.relpath(root, base_wav_dir)  # e.g., 84/121123
#             speaker_id, chapter_id = rel_folder.split(os.sep)
#             file_id = file.replace(".wav", "")  # e.g., 84-121123-0000

#             # Load transcript
#             transcript_file = os.path.join(
#                 base_transcript_dir, speaker_id, chapter_id, f"{speaker_id}-{chapter_id}.trans.txt"
#             )

#             if not os.path.exists(transcript_file):
#                 print(f"Transcript not found for {wav_path}")
#                 continue

#             with open(transcript_file, "r", encoding="utf-8") as f:
#                 transcript_lines = {
#                     line.split(" ", 1)[0]: line.split(" ", 1)[1].strip()
#                     for line in f if " " in line
#                 }

#             transcript = transcript_lines.get(file_id, None)
#             if transcript:
#                 data.append({
#                     "file_id": file_id,
#                     "wav_path": wav_path,
#                     "transcript": transcript
#                 })

# # 🔍 Preview
# for d in data[:3]:
#     print(f"{d['file_id']}\nTranscript: {d['transcript']}\nAudio: {d['wav_path']}\n")
import os

# Directories
wav_dir = r"Data\Train\librispeech_asr\dev-clean\converted_wav\84\121123"
transcript_file = r"Data\Train\librispeech_asr\dev-clean\84\121123\84-121123.trans.txt"

# Load transcripts into a dict
transcript_lines = {}
with open(transcript_file, "r", encoding="utf-8") as f:
    for line in f:
        parts = line.strip().split(" ", 1)
        if len(parts) == 2:
            utt_id, text = parts
            transcript_lines[utt_id] = text

# Pair wav files with their transcripts
data = []
for file in os.listdir(wav_dir):
    if file.endswith(".wav"):
        file_id = file.replace(".wav", "")  # e.g., 84-121123-0000
        wav_path = os.path.join(wav_dir, file)

        transcript = transcript_lines.get(file_id, None)
        if transcript:
            data.append({
                "file_id": file_id,
                "wav_path": wav_path,
                "transcript": transcript
            })

# 🔍 Preview
for d in data:
    print(f"{d['file_id']}\nTranscript: {d['transcript']}\nAudio: {d['wav_path']}\n")
```
